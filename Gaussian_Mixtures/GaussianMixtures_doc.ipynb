# Simulating and Visualizing Gaussian Mixtures

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from GaussianMixture import GaussianMixture
```

We begin by initializing the parameters for our Gaussian mixture. The parameters required include `pi`, which are the weights allocated to a given Gaussian within a mixture, `centroids` which are the means of each of the Gaussian in the mixture, while `covmat` are the diagonals of each covariance matrix. For the time being, we are assuming off-diagonal elements of the covariance matrix are zero. 

```python
# mixture coefficients
pi = np.array([20, 34, 32, 45, 40, 29, 70])
pi = pi/pi.sum()

# centroids
centroids = np.array([
					 [0, 0],
					 [-15, 5],
					 [-5, -9],
					 [4, -3],
					 [8, 12],
					 [2, 2],
					 [15, -1]])

# covariance matrix
covmat = np.array([
					 [3, 4],
					 [3, 2.5],
					 [2, 3],
					 [14, 2],
					 [4, 3],
					 [3, 4],
					 [4, 13]])
```

We now initialize the `GaussianMixture` class with the parameters above. From this class, we can call the `simulate_gm` method for a number of samples.

```python
# initialize the gaussian mixture class
gm = GaussianMixture(pi, centroids, covmat)

# simulate samples from the mixture
gm_sim = gm.simulate_gm(n_samples = 20)
```

Calling the method `pdf_gm` will give you the probability density of a particular input. You can input a single input or an array of inputs matching the dimension of the mixture.

```python
# call the pdf of the mixture
pdf_out = gm.pdf_gm(gm_sim)
display(pdf_out)
```

The `surface_gm` method in the `GaussianMixture` class plots the 3D visualization of the Gaussian mixture over a specified range and grid fineness.

```python
# surface plot of bivariate mixture
gm.surface_gm(ranges=[-25,25], fineness=.1, filename="testplot.png")
```

We can also visualize a pairwise contour/scatter plot where the diagonals are histograms by calling the `pairs_gm` method. The input can be any multivariate mixture in an NxD array.

```python
# pairwise contour plot of data
gm.pairs_gm(gm_sim, ranges=[-23, 23], filename="testplot2.png")
```

Finally, to save mixture parameter results and the true first and second moments, we can call the `write_gm` and `write_params` methods.

```python
# write simulation output to file
gm.write_gm(gm_sim, filename="sim_output.csv")

# write true parameters of mixture to file
gm.write_params(filename="true_params.csv")
```
